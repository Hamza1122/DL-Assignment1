```
import requests as re
from bs4 import BeautifulSoup
import json
import csv
import os
from datetime import datetime
import yfinance as yf
import time
import pandas as pd
re=re.Session()
TM=datetime
CSV_LOAD_FILE='Trading View Stocksa.csv'
CSV_LOG_FILE='LOG.csv'
YAHOO_FINANCE_URL='https://finance.yahoo.com/quote/'
CHECK_URL_LINK='https://query1.finance.yahoo.com/v7/finance/quote?symbols='
SUMMARY_CSV_HEADER=['Ticker','Quote Type','Previous Close', 'Open', 'Bid', 'Ask', "Day's Range", '52 Week Range', 'Volume', 'Avg. Volume', 'Market Cap', 'Beta (3Y Monthly)', 'PE Ratio (TTM)', 'EPS (TTM)', 'Earnings Date', 'Forward Dividend & Yield', 'Ex-Dividend Date', '1y Target Est','Inception Date', 'Net Assets', 'Expense Ratio (net)', 'Yield', 'NAV', 'YTD Daily Total Return','YTD Return', 'Sustainability Rating', '5y Average Return', 'Average for Category', 'Last Dividend', 'Morningstar Rating', 'Last Cap Gain', 'Morningstar Risk Rating', 'Category', 'Holdings Turnover','Beta (5Y Monthly)']
SUMMARY_CSV_FILE_NAME='Stocks Summary.csv'
listOfTickers=[]
downloadedTickers=[]
requestHeader={
  'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3',
  'accept-language': 'en-US,en;q=0.9',
  'cache-control': 'no-cache',
  'pragma': 'no-cache',
  'upgrade-insecure-requests': '1',
  'user-agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36',
}
```

```

def init(overwrite=False):
  if not os.path.exists(SUMMARY_CSV_FILE_NAME) or overwrite:
    with open(SUMMARY_CSV_FILE_NAME,'w',newline='') as csvfile:
      filewriter = csv.DictWriter(csvfile, fieldnames=SUMMARY_CSV_HEADER)
      filewriter.writeheader()

def checkCreatePath(directory):
    directory=directory.replace('/','')
    if not os.path.exists(directory):
        os.makedirs(directory)
        print("[>] "+directory+" Folder Created.")

def initEmptyFile(FILE_NAME):
  open(FILE_NAME,'w').close()

def writeRowForSummary(rowDict):
  with open(SUMMARY_CSV_FILE_NAME,'a',newline='',encoding='utf-8-sig') as csvfile:
    filewriter = csv.DictWriter(csvfile, fieldnames=SUMMARY_CSV_HEADER)
    filewriter.writerow(rowDict)

```

```
def writeStatementRow(CSV_FILE_NAME,rowAsList):
  with open('Revised Ticker Files/'+CSV_FILE_NAME, 'a',newline='',encoding='utf-8-sig') as csvfile:
      wr = csv.writer(csvfile)
      wr.writerow(rowAsList)

def getTickerInformation(ticker):
  pageJson=re.get(CHECK_URL_LINK+ticker).json()
  result=pageJson['quoteResponse']['result']
  # If result in JSon is empty it will pass the result T/F flag and result JSON
  if result==[] or result is None:return False,result
  else:
    if 'longName' not in result[0]:
      result[0]['longName']=''
    return True,result[0]
  # Only one ticker is passed thats why only first result is needed

def getTickerSummary(dataDict):
  ticker=dataDict['Ticker']
  print('  # '+ticker+'\'s Stock Summary')
  while True:
    page=re.get(YAHOO_FINANCE_URL+ticker,headers=requestHeader)
    pageParser=BeautifulSoup(page.content,'html.parser')
    try:
      summaryDiv=pageParser.find('div',{'id':'quote-summary'})
      leftTableTr=summaryDiv.find('div',{'data-test':'left-summary-table'}).find('table').findAll('tr')
      rightTableTr=summaryDiv.find('div',{'data-test':'right-summary-table'}).find('table').findAll('tr')
    except AttributeError as e:
      print(' . '+str(e))
      if 'lookup' in page.url:
        return False
      input("Please have a look here.getTickerSummary ")
      continue
    break
  for row in leftTableTr+rightTableTr:
    name,value=row.findAll('td')
    dataDict[name.text.strip()]=value.text.strip()
  writeRowForSummary(dataDict)
  return True



def getHistoricalData(ticker):
  print('    # '+ticker+'\'s Historical Data')
  yf_fileName='Revised Ticker Files/'+ticker+'_Historical Data.csv'
  initEmptyFile(yf_fileName)
  tick = yf.Ticker(ticker)
  try:
    df=tick.history(period="max")
    df.to_csv(yf_fileName)
  except ValueError as e:
    if 'No data found, symbol may be delisted' in str(e):
        print('> '+ticker+' Historical Data Not Found')
        return
    df=tick.history(period="5Y")
    df.to_csv(yf_fileName)
    print(' [!] Downloading for 5 Years')
```

```
def loadTickers():
  print('[#] Started Loading Files.')
  try:
    with open(CSV_LOAD_FILE,'r', newline='',encoding='utf-8-sig') as f:
      reader = csv.reader(f)
      next(reader)
      try:
        for row in reader:
            listOfTickers.append(row[4].strip())
      except UnicodeDecodeError :
        pass
  except FileNotFoundError as f:
    print('[!] '+CSV_LOAD_FILE+' File Not Found')
  
  try:
    with open(CSV_LOG_FILE,'r', newline='',encoding='utf-8-sig') as f:
      reader = csv.reader(f)
      for row in reader:
        
        downloadedTickers.append(row[0].strip())  
  except FileNotFoundError as f:
    print('[!] LOG File will be created in next iteration')

  print("[>] Total Tickers  = ",len(listOfTickers))
  print("[>] Total Download = ",len(downloadedTickers))
  #print('[=] Complete perc  = %.2f' % ((len(downloadedTickers)/len(listOfTickers))*100))
  #print()

def reportError(ticker,msg,json):
    with open('logInformation','a',newline='',encoding='utf-8-sig') as csvfile:
      csvfile.write(datetime.now().strftime("%d/%m/%Y %I:%M %p")+','+ticker+','+str(msg)+','+str(json)+'\n')
    
def storeLog(ticker,jsonResponse,isTicker=True):
  dataDict={
      'Ticker':ticker,
      'Date':datetime.now().strftime("%d-%m-%Y"),
      'Time':datetime.now().strftime("%I:%M %p")
    }
  if not isTicker:
    dataDict['Status']='Ticker not Found'
  else:
    if 'longName' in jsonResponse:
      dataDict['Name']=jsonResponse['longName']

    dataDict['Status']='Available'
    dataDict['Type']=jsonResponse['quoteType']

  with open(CSV_LOG_FILE,'a',newline='',encoding='utf-8-sig') as csvfile:
    filewriter = csv.DictWriter(csvfile, fieldnames=['Ticker','Status','Name','Type','Date','Time'])
    filewriter.writerow(dataDict)
```

```

def searchTicker(ticker):
  try:
    dataJson=re.get('https://query2.finance.yahoo.com/v1/finance/search?q='+ticker).json()
    #JSON DEcode error fixed here because of encodeed data sent by yahoo fnance
  except json.decoder.JSONDecodeError:
    return 'N/A'
  if dataJson['quotes']==[]:
    return 'N/A'
  else:
    return dataJson['quotes'][0]['symbol']
  
checkCreatePath('Revised Ticker Files')
init()
loadTickers()
quoteType=['EQUITY','MUTUALFUND','ETF','CRYPTOCURRENCY']
for ticker in listOfTickers:
  try:
    if ticker in downloadedTickers:continue
    
    isTicker,responseJson=getTickerInformation(ticker)
    if not isTicker:
      tick=searchTicker(ticker)
      if tick=='N/A':
        print(ticker,' is not ticker Symbol')
        storeLog(ticker,responseJson,False)
        continue
      else:
        print(ticker+' is '+tick)
        ticker=tick
        isTicker,responseJson=getTickerInformation(ticker)
    if responseJson["quoteType"] not in quoteType:
      print(responseJson["quoteType"])
      storeLog(ticker,'Differed QuoteType'+responseJson["quoteType"],False)
      continue
    print("> "+ticker.ljust(10)+'   '+responseJson["quoteType"])
    #Fixed the issue of MUTUALFUNDS new header
    try:
      isCorrect=getTickerSummary({'Ticker':ticker,'Quote Type':responseJson["quoteType"]})
      if not isCorrect:continue
    except ValueError as e:
      print(e)
    getHistoricalData(ticker)
    re.cookies.clear()
    storeLog(ticker,responseJson)
    print('---------------------------------')
  except (KeyError,AttributeError) as e:
    print(e)
    storeLog(ticker,responseJson)
    print(ticker+' , Something wrong Happened.')
print('\n\n\n==============Scraping Completed===========')
# ETF AAAU,VOO
# https://query1.finance.yahoo.com/v10/finance/quoteSummary/FB?formatted=true&crumb=g6cgBsM1lX1&lang=en-US&region=US&modules=price%2CsummaryDetail%2CpageViews&corsDomain=finance.yahoo.com
# https://query1.finance.yahoo.com/v8/finance/chart/AAPL
# https://query2.finance.yahoo.com/v1/finance/search?q=FB
# https://query1.finance.yahoo.com/v1/finance/search?q=FB&quotesCount=6&newsCount=4&enableFuzzyQuery=false&quotesQueryId=tss_match_phrase_query&multiQuoteQueryId=multi_quote_single_token_query&newsQueryId=news_cie_vespa&enableCb=true&enableNavLinks=true&vespaNewsTimeoutMs=600
# for finding the ticker
# https://query1.finance.yahoo.com/v7/finance/quote?symbols=TLS.AX
```
