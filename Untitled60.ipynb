```
import pandas as pd                   #libraries
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import cross_val_score
import tensorflow as tf
from keras.layers import Dense, Dropout, Activation
from keras.models import Sequential
from keras.layers import Dense, Embedding, GlobalMaxPool1D, Dropout
from keras.optimizers import Adam
from keras.optimizers import SGD
```

```
df=pd.read_csv('fyp.csv')   #loading the dataset
```

```
df.head(10)   #displaying the first 10 rows of a dataset
```

```
df.isnull().sum()   #checking if there any null values
```

```
del df['Floor']     #removing the columns
del df['Run']       ##
del df['XandY']
#del df['XCoordinate']
#del df['YCoordinate']
```

```
Y=df[['XCoordinate','YCoordinate']].values                 #assigning labels(column) to a Y 
X=df.drop(['XCoordinate','YCoordinate'],axis=1).values     #assigning all data excep labels column
```

```
X=np.array(X)
Y=np.array(Y)
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)   #randomly splitting data into train and test
```

```
X_train.shape     #size of X_train data
```

```
y_train.shape
```

```
model = Sequential()                                     #Applying Multi-Layer Perceptron
model.add(Dense(510,input_dim=223))       #Declaring input dimension,neurons and activation function 
model.add(Dropout(0.5))                                      #using drop out to overcome the overfitting
model.add(Dense(64))                    
model.add(Dropout(0.5))
model.add(Dense(2))                     
model.compile(loss='mean_absolute_error', optimizer='Adam')   #mean_absolute calculating the error of a model and 
model.fit(X_train, y_train,epochs=600,batch_size=60)          #passing parameters into a model to train the neural network
```

```
print('\n# Evaluate on test data')
results = model.evaluate(X_test,y_test, batch_size=128)  #evaluating the result on test data
print('test loss:', results)

```

```
# Dump weights
for layerNum, layer in enumerate(model.layers):
    weights = layer.get_weights()[0]
    biases = layer.get_weights()[1]
    
    for toNeuronNum, bias in enumerate(biases):
        print(layerNum,'B -> L',layerNum+1,'N',toNeuronNum,bias)
    
    for fromNeuronNum, wgt in enumerate(weights):
        for toNeuronNum, wgt2 in enumerate(wgt):
            print('L',layerNum,'N',fromNeuronNum,'-> L',layerNum+1,'N',toNeuronNum,wgt2)
```

```

print('\n# Generate predictions for 3 samples')
predictions = model.predict(X_test)
print('predictions shape:', predictions.shape)
```
